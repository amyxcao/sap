```r
library(ggplot2)

age<-ggplot(data = test, aes(x = Age, fill = factor(Customer_Churn))) +
  geom_histogram(binwidth = 5, position = "dodge") +
  labs(title = "Distribution of Age by Churn",
       x = "Age",
       y = "Count") +
  scale_fill_manual(values = c("#0072B2", "#D55E00"), labels = c("No Churn", "Churn"))
age
```

```r
#correlation between the age of the customers and churn.
```

```r
ai_sat<-ggplot(data = test, aes(x = factor(Customer_Churn), y = Satisfaction_with_AI_Services, fill = factor(Customer_Churn))) +
  geom_boxplot() +
  labs(title = "Satisfaction with AI Services by Churn",
       x = "Churn",
       y = "Satisfaction") +
  scale_fill_manual(values = c("#0072B2", "#D55E00"), labels = c("No Churn", "Churn"))
ai_sat
```

```r
#the distribution of satisfaction ratings with AI services among customers who churned and those who didn't.
```

```r
usage<-ggplot(data = test, aes(x = AI_Interaction_Level, y = Overall_Usage_Frequency, color = factor(Customer_Churn))) +
  geom_point() +
  labs(title = "AI Interaction Level vs. Overall Usage Frequency",
       x = "AI Interaction Level",
       y = "Overall Usage Frequency (days in a month)") +
  scale_color_manual(values = c("#0072B2", "#D55E00"), labels = c("No Churn", "Churn"))
usage
```

```r
#the relationship between the frequency and depth of AI interactions and overall usage frequency.
```

```r
response_time<-ggplot(data = test, aes(x = AI_Response_Time, fill = factor(Customer_Churn))) +
  geom_bar(position = "dodge") +
  labs(title = "AI Response Time by Churn",
       x = "AI Response Time (categorical)",
       y = "Count") +
  scale_fill_manual(values = c("#0072B2", "#D55E00"), labels = c("No Churn", "Churn"))
response_time
```

```r
#[distribution of AI response time categoies among customers who churned and those who didnt]
```

```r
usage_pattern<-ggplot(data = test, aes(x = factor(Change_in_Usage_Patterns), fill = factor(Customer_Churn))) +
  geom_bar(position = "stack") +
  labs(title = "Change in Usage Patterns by Churn",
       x = "Change in Usage Patterns",
       y = "Count") +
  scale_fill_manual(values = c("#0072B2", "#D55E00"), labels = c("No Churn", "Churn"))
usage_pattern
```

```r
#distribution of change in usage patterns among customers who churned and those who didn't.
```
